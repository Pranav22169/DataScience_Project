```python
#Part 3(Sentiment Analysis and Normality Testing) for Ola dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot  # Importing probplot here
import numpy as np

# Load the dataset
df = pd.read_excel(r"C:\Users\prabi\OneDrive\Desktop\V sem\FDS_project\Datasets\Ola.xlsx")

# Check the unique values in the 'Outcome' column
print("Unique values in 'Outcome':", df['Outcome'].unique())

# Step 1: Calculate sentiment metrics
# Convert the Outcome column to lowercase before mapping
df['sentiment'] = df['Outcome'].str.lower().map({'positive': 1, 'neutral': 0, 'negative': -1})

# Check for any NaN values after mapping
print("Sentiment values after mapping:\n", df['sentiment'].head())
print("Number of NaN values in sentiment:", df['sentiment'].isna().sum())

# Proceed only if there are valid sentiment values
if df['sentiment'].notna().any():
    # Calculate metrics
    mean_score = df['sentiment'].mean()
    std_dev = df['sentiment'].std(ddof=1)
    median = df['sentiment'].median()

    print(f"Mean Sentiment Score: {mean_score:.4f}")
    print(f"Standard Deviation: {std_dev:.4f}")
    print(f"Median: {median:.4f}")

    # Step 2: Check normality of the sentiment scores
    sentiment_data = df['sentiment'].dropna()

    if len(sentiment_data) >= 3:
        stat, p_value = shapiro(sentiment_data)
        print(f"Shapiro-Wilk Test Statistic: {stat:.4f}, P-Value: {p_value:.4f}")

        if p_value > 0.05:
            print("Data is normally distributed (Fail to reject H0).")
        else:
            print("Data is not normally distributed (Reject H0).")

        # Q-Q Plot
        plt.figure(figsize=(8, 6))
        probplot(sentiment_data, dist="norm", plot=plt)  # Using the imported probplot
        plt.title('Q-Q Plot of Sentiment Scores')
        plt.grid()
        plt.show()

        # Histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(sentiment_data, kde=True)
        plt.title('Histogram of Sentiment Scores')
        plt.grid()
        plt.show()
    else:
        print("Not enough data points for normality test.")
else:
    print("No valid sentiment data available.")
```
